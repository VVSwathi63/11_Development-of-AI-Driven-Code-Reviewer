```python
student_code = """
x=10
if(x>5):print(x)
"""
```

```python
import ast

try:
    tree = ast.parse(student_code)
    print("Code parsed successfully")
except SyntaxError as e:
    print("Syntax Error:", e)
```

```python
print(ast.dump(tree, indent=4))
```

```python
import autopep8

formatted_code = autopep8.fix_code(student_code)
print(formatted_code)
```

```python
import black

formatted_code = black.format_str(student_code, mode=black.FileMode())
print(formatted_code)
```
